2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

# a)
outer_product_np= np.outer (u, v)
print ("oter product with np:\n ", outer_product_np)

# b)
# Initialize an empty matrix to store the outer product ( di dimensioni len(u) e len(v))
outer_product_loop = np.zeros((len(u), len(v)))
# Calculate the outer product using a nested for loop
for i in range(len(u)):
    for j in range(len(v)):
        outer_product_loop[i, j] = u[i] * v[j]

print("Outer product using nested for loop:\n", outer_product_loop)

# c)
# Reshape u and v to have the shape (4, 1) and (1, 4) respectively. 
u_reshaped = u[:, np.newaxis]   #Acheived by adding a new axis along the second dimention
v_reshaped = v[np.newaxis, :]   # achieved by adding a new axis along the first dimension (axis with index 0).
# Calculate the outer product using broadcasting
outer_product_broadcast = u_reshaped * v_reshaped

print("Outer product using numpy broadcasting:\n", outer_product_broadcast)

#(mentre se faccio
# product= v * u  mi moltiplica elemento per elemento )
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
def sieve(N):
    mask= np.ones(N+1, dtype= bool) # from now on True= prime number
    mask[0]= mask[1]= False # 0 e 1 non sono numeri primi

    for i in range (2, N+1):
        if mask[i]== True:
            mask[i*i::i]= False
    prime_numbers= np.arange(N+1)[mask]
    return prime_numbers

N=99
print (sieve(N))
````

````python
import timeit
%timeit sieve(N)
%timeit sieve(N*2)
%timeit sieve(N*4)
%timeit sieve(N*6)
%timeit sieve(N*10)
````

````python
def sieve_Eratosthenes(N):
    mask = np.ones(N+1, dtype=bool)
    mask[0]=mask[1]=False
    for i in range(2, int(np.sqrt(N))+1):
        if mask[i]==True:
            mask[i+i::i]=False
    prime_numbers = np.arange(N+1)[mask]
    return prime_numbers

N=99
print(sieve_Eratosthenes(N))

%timeit sieve_Eratosthenes(N)
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
walkers= 1000
steps= 200
m = np.random.randint(2, size=(walkers,steps)) # array made of 1 and 0 (2= high value exclusive)
m[m==0] = -1 # array made of 1 and -1

walker_distance  = np.cumsum(m, axis=1) # distance traveled by each walker
wd2 = walker_distance**2                
mean_wd2 = np.mean(wd2, axis=0)  #mean squared distance for each step across all walkers.

plt.plot(np.sqrt(mean_wd2), label='simulated')
plt.plot(np.sqrt(np.arange(steps)), label='theoretical') # in theory d=sqrt(200)
plt.xlabel("steps")
plt.ylabel("mean walking distance")
plt.legend()
plt.show()
````

Do you feel there is some evident correlation here? [Studies](https://www.enr.gov.nt.ca/en/services/lynx/lynx-snowshoe-hare-cycle) tend to believe so.

````python
# copiato velocemente, leggo beben il codice
import numpy as np
data = np.loadtxt('populations.txt', skiprows=1)

# Create arrays for each column
year = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

# Plot the populations over the years
plt.figure(figsize=(10, 6))
plt.plot(year, hares, label='Hares', marker='o')
plt.plot(year, lynxes, label='Lynxes', marker='o')
plt.plot(year, carrots, label='Carrots', marker='o')

plt.title('Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()

# Compute statistical properties
mean_hares = np.mean(hares)
std_hares = np.std(hares)
mean_carrots = np.mean(carrots)

std_hares = np.std(hares)
std_lynxes = np.std(lynxes)
std_carrots = np.std(carrots)

correlation_hares_lynxes = np.corrcoef(hares, lynxes)[0, 1]
correlation_hares_carrots = np.corrcoef(hares, carrots)[0, 1]
correlation_lynxes_carrots = np.corrcoef(lynxes, carrots)[0, 1]

# Display the statistical properties
print(f'Mean Hares: {mean_hares}')
print(f'Mean Lynxes: {mean_lynxes}')
print(f'Mean Carrots: {mean_carrots}')
print(f'Std Hares: {std_hares}')
print(f'Std Lynxes: {std_lynxes}')
print(f'Std Carrots: {std_carrots}')
print(f'Correlation Hares-Lynxes: {correlation_hares_lynxes}')
print(f'Correlation Hares-Carrots: {correlation_hares_carrots}')
print(f'Correlation Lynxes-Carrots: {correlation_lynxes_carrots}')

# Determine which species has the highest population each year
max_population_species = np.argmax(data[:, 1:], axis=1)

# Map the index to species
species_mapping = {0: 'Hares', 1: 'Lynxes', 2: 'Carrots'}
max_species_each_year = [species_mapping[idx] for idx in max_population_species]

print(f"Species with the highest population each year: {max_species_each_year}")
````
